```python
import math 
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
def next_permutation(L):
    n = len(L)
    i = n - 2
    while i >= 0 and L[i] >= L[i+1]:
        i -= 1
    if i == -1:
        return False
    j = i + 1
    while j < n and L[j] > L[i]:
        j += 1
    j -= 1
    L[i], L[j] = L[j], L[i]
    left = i + 1
    right = n - 1
    while left < right:
        L[left], L[right] = L[right], L[left]
        left += 1
        right -= 1
             
    return True

def travllingSalesmanProblem(graph, s):
    list = []
    for x in range(0, 5):
        if x != s:
            list.append(x)
            
    min_path = math.inf
    while True:
        print(0, end =" -> ") 
        print(*list, sep = ' -> ',end =" -> ")
        print(0, end = "      ")
        current_pathweight = 0
        k = s
        for x in range(0,len(list)):
            current_pathweight = current_pathweight + graph.item((k, list[x]))
            k = list[x]
        current_pathweight = current_pathweight + graph.item((k, s))
        print(current_pathweight)
        min_path = min(min_path, current_pathweight)
        if not next_permutation(list):
            break
    
    return min_path
 
 
 
if __name__ == "__main__":
    graph = np.matrix([[0, 10, 8, 9, 7], [10, 0, 10, 5, 6], [8, 10, 0, 8, 9], [9, 5, 8, 0, 6], [7, 6, 9, 6, 0]])
    G = nx.from_numpy_matrix(graph)
    G.add_edges_from([(1, 2), (2, 1), (2, 3)])
    nx.draw(G)
    plt.show()
    s = 0
    md = travllingSalesmanProblem(graph, s)
    print("Minimum Distance to cover all cities = ",end = " ")
    print(md)
```
